```python
import pandas as pd
df=pd.read_csv("../data/netflix_titles.csv")
df.head()
```

```python
print("Rows, Colums:", df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe(include="all")
```

```python
df["director"] = df["director"].fillna("Unknown")

df["cast"] = df["cast"].fillna("Unknown")

df["country"] = df["country"].fillna("Unknown")

df["rating"] = df["rating"].fillna(
    df["rating"].mode()[0]
)

df["duration"] = df["duration"].fillna(
    df["duration"].mode()[0]
)

df.dropna(subset=["date_added"], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])

df["year_added"] = df["date_added"].dt.year

df["month_added"] = df["date_added"].dt.month

df.head()
```

```python
# Convert date column safely
df["date_added"] = pd.to_datetime(
    df["date_added"],
    errors="coerce",
    format="mixed"
)

# Remove any rows that still have invalid dates
df.dropna(subset=["date_added"], inplace=True)

# Create new features
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month

print("Feature Engineering Complete!")
```

```python
df[["date_added","year_added","month_added"]].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))

sns.countplot(x="type", data=df)

plt.title("Movies vs TV Shows on Netflix")

plt.savefig("../visualizations/content_type.png")

plt.show()
```

```python
plt.figure(figsize=(10,5))

df["country"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.tight_layout()

plt.savefig("../visualizations/country.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    y="rating",
    data=df,
    order=df["rating"].value_counts().index
)

plt.title("Content Rating Distribution")

plt.savefig("../visualizations/ratings.png")

plt.show()
```

```python
plt.figure(figsize=(10,5))

df["year_added"].value_counts().sort_index().plot(
    kind="line",
    marker="o"
)

plt.title("Netflix Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.grid(True)

plt.savefig("../visualizations/yearly_growth.png")

plt.show()
```
